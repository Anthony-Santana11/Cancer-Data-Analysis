```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import os 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cancer.csv")
```

```python
desc_stats = df.describe()
desc_stats
```

```python
plt.figure(figsize=(14,8))
sns.set(style="whitegrid")
```

```python
# Bars Graphics for development risk em females and males

plt.subplot(2,1,1)
sns.barplot(x='Cancer Type' ,y='Male Risk Development Percentage', data=df,color='blue', alpha = 0.6, label = 'Homens')
sns.barplot(x='Cancer Type' , y='Female Risk Development Percentage', data=df , color = 'red' , alpha = 0.6 , label = 'Mulheres')
plt.title('Risco de Desenvolvimento de Câncer (%)')
plt.ylabel('Percentual de Risco de Desenvolvimento')
plt.xlabel('Tipo de Câncer')
plt.xticks(rotation = 90)
plt.legend()
```

```python

# Dying risk for males and females 

sns.barplot(x='Cancer Type' ,y='Male Risk Development Percentage', data=df,color='blue', alpha = 0.6, label = 'Homens')
sns.barplot(x='Cancer Type' , y='Female Risk Development Percentage', data=df , color = 'red' , alpha = 0.6 , label = 'Mulheres')
plt.title('Risco de Morte por Câncer (%)')
plt.ylabel('Percentual de Risco de Morte')
plt.xlabel('Tipo de Câncer')
plt.xticks(rotation = 90)
plt.legend()
plt.show()

# Show desc stats
desc_stats
```
